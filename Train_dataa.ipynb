# clean the ['Arrival Delay in Minutes'] column

```python
impute=int(np.round(data['Arrival Delay in Minutes'].mean()))
data['Arrival Delay in Minutes'].fillna(impute,inplace=True)
data['Arrival Delay in Minutes'].isna().sum()
```

# Drop Useless Columns

```python
data.drop(['Unnamed: 0'],axis=1)
data.drop(data[data['Seat comfort'] == 0].index ,inplace = True) 
```

# Showing Ages

```python
bins = np.arange(5, 85, 5)
plt.hist(data.Age, bins= bins ,edgecolor = 'black', color ='#921A40')

plt.xlim(5, 85)
plt.xticks(bins)
plt.grid(True, linestyle='-', color='black', alpha=0.4)
plt.title('Age Distribution')
plt.xlabel('Age')
```

# What is the most common gender among travelers, male or female?


```python
gender_counts = data['Gender'].value_counts()
color=['red','blue']
plt.pie(gender_counts, labels=gender_counts.index,colors=color ,autopct='%1.1f%%')
plt.title('Distribution of Gender in Travel')
```

# How does the distribution of gender compare between business and personal travelers?

```python
business_travel_data = data[data['Type of Travel'] == 'Business travel']
Personal_Travel_data = data[data['Type of Travel'] == 'Personal Travel']

fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(12, 5))
business_travel_data['Gender'].value_counts().plot(kind='bar',edgecolor='red',cmap='summer' ,ax=ax1)
ax1.set_xlabel('Gender')
ax1.set_ylabel('Count')
ax1.set_title('Gender of Business Travelers')

# Personal Travel
Personal_Travel_data['Gender'].value_counts().plot(kind='bar',edgecolor='yellow',cmap='summer', ax=ax2)
ax2.set_xlabel('Gender')
ax2.set_ylabel('Count')
ax2.set_title('Gender of Personal Travelers')

plt.tight_layout()
plt.show()
```

# What is the number of passengers who rated the seats as comfortable in each class?

```python

shifts = np.arange(1,6)
width = .25
plt.bar(shifts   , data.loc[data['Class'] == 'Eco','Seat comfort'].value_counts().sort_index(), width =width ,label = 'Eco Class')
plt.bar(shifts -width , data.loc[data['Class'] == 'Eco Plus','Seat comfort'].value_counts().sort_index(), width =width ,label = 'Eco plus Class')
plt.bar(shifts + width , data.loc[data['Class'] == 'Business','Seat comfort'].value_counts().sort_index(), width =width ,label = 'Business Class')

plt.grid(True, linestyle='-', color='black', alpha=0.4)
plt.legend()
plt.xlabel('Seat comfort Rate')
plt.ylabel('Number of passengers')
plt.title('Seat comfort rate for the 3 classes')
```

# What number of passengers were satisfied with the time delay and wifi service

```python
shifts = np.arange(2)
width = 0.25

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

filter1 = data.loc[data['Inflight wifi service'] < 4, 'satisfaction'].value_counts().reindex(['neutral or dissatisfied', 'satisfied'], fill_value=0)
filter2 = data.loc[data['Inflight wifi service'] >= 4, 'satisfaction'].value_counts().reindex(['neutral or dissatisfied', 'satisfied'], fill_value=0)

ax1.bar(shifts, filter1, width=width, label='Wifi service rate < 4')
ax1.bar(shifts + width, filter2, width=width, label='Wifi service rate >= 4')

ax1.set_xticks(ticks=shifts, labels=['Unsatisfied', 'Satisfied'], size=12)
ax1.legend(loc='upper right')
ax1.set_title('Passengers Satisfaction Compared to Wifi Service Rate')
ax1.set_ylabel('Number of Passengers', size=14)
ax1.set_xlim(-0.5, 2.5)
ax1.grid(True, alpha=0.7)

mean_value = data['Departure Delay in Minutes'].mean()
filter3 = data.loc[data['Departure Delay in Minutes'] <= mean_value, 'satisfaction'].value_counts().reindex(['neutral or dissatisfied', 'satisfied'], fill_value=0)
filter4 = data.loc[data['Departure Delay in Minutes'] > mean_value, 'satisfaction'].value_counts().reindex(['neutral or dissatisfied', 'satisfied'], fill_value=0)

ax2.bar(shifts, filter3, width=width, label='Departure delay time <= mean delay')
ax2.bar(shifts + width, filter4, width=width, label='Departure delay time > mean delay')

ax2.set_xticks(ticks=shifts, labels=['Unsatisfied', 'Satisfied'], size=12)
ax2.legend(loc='upper right')
ax2.set_title('Passengers Satisfaction Compared to the Average Departure Delay')
ax2.set_xlim(-0.5, 2.5)
ax2.grid(True, alpha=0.7)

plt.tight_layout()
plt.show()
```

# What is the number of satisfied and unsatisfied responses among older people

```python
elderly = data[data['Age'] >= 65]['satisfaction'].value_counts()


plt.pie(elderly ,labels = elderly.index, autopct = '%1.1f%%' ,textprops= {'fontsize': 14},
        wedgeprops = {'edgecolor' : 'black'}, startangle = 40 ,shadow = True)
```

# Summary:

- We found that the proportion of males is 49.3%, while the proportion of females is 50.7%.
- We found that the number of males exceeds the number of females in both business and personal travel categories
- We found that the Business class received the highest number of votes, with a rating of 4.
- We found that passengers are more satisfied with a wifi service rating greater than 4. Additionally, satisfaction increases as the departure time delay decreases.
- We found that older individuals are generally less satisfied and more often unsatisfied.
